### Classification step

Here, we use a tree based model that is known to be very effective for this kind of tasks. Hence, we don't need to handle multicolinearity, outliers and normalization fo the data. H

```python
n = 1500 

df.loc[df['price_label'] == 5] = 4
df_subsampled0 = df[df['price_label'] == 4]
df_subsampled1 = df[df['price_label'] == 0].sample(n=n, random_state=0)
df_subsampled2 = df[df['price_label'] == 1].sample(n=n, random_state=0)
df_subsampled3 = df[df['price_label'] == 2].sample(n=n, random_state=0)
df_subsampled4 = df[df['price_label'] == 3].sample(n=n, random_state=0)

df_concatenated = pd.concat([df_subsampled0, df_subsampled1, df_subsampled2, df_subsampled3, df_subsampled4])
df_concatenated['price_label'].value_counts()
```

```python
from sklearn.model_selection import train_test_split

# Define the target
X = df_concatenated.drop(['price_label', 'average_price', 'avg_selling_price'],axis=1)
y = df_concatenated.price_label

# Split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)
```



### Balance the classes 

First we have to manually subsample the majority class and then we can use an algorithm.

ADASYN (Adaptive Synthetic Sampling) is a type of oversampling technique for imbalanced datasets. Unlike traditional oversampling techniques like SMOTE which oversamples the minority class with a fixed ratio, ADASYN dynamically adjusts the oversampling ratio for minority samples based on their difficulty in being classified.

The idea behind ADASYN is to oversample the minority samples that are difficult to be classified correctly. These samples are usually located near the decision boundary between classes and may cause the classifier to be biased towards the majority class. By oversampling these samples, the classifier can be trained to better handle the minority class.

In ADASYN, a density distribution is calculated for the minority samples based on their distances to their k-nearest neighbors. The samples with higher density are more likely to be oversampled. This way, the oversampling ratio is adjusted based on the difficulty of the minority samples, rather than a fixed ratio.

```python
y_train.value_counts()
```

```python
from imblearn.over_sampling import ADASYN

adasyn = ADASYN(sampling_strategy='minority', random_state=0)
X_resampled, y_resampled = adasyn.fit_resample(X_train, y_train)
y_resampled.value_counts()
```

### Modeling with a tree based model 

From the exploration we did thanks to Pycaret, we can confidently choose a tree based model for this dataset. A boosted tree in particular. Those are really strong models capable of modeling non linear relationship in the data. I choose LGBM from Microsoft because its overall the best performing model on this dataset and because it is lightweight and very reliable in production. It is important to find a sweet spot between performance and cloud costs.Here we do a proper training using the lightgbm library.

```python
import lightgbm as lgb
from sklearn.model_selection import RandomizedSearchCV


clf = lgb.LGBMClassifier()
        
# Random Search for Hyperparameters
param_grid = {
    'boosting_type': ['gbdt'],
    'objective': ['binary'],
    'num_leaves': [32, 64, 128, 256],
    'learning_rate': [0.05, 0.1, 0.15, 0.2, 0.5],
    'min_child_samples': [10, 20, 30, 40],
    'n_estimators': [50, 100, 200, 500, 750],
    'verbose': [-1]
    }

clf_random = RandomizedSearchCV(estimator = clf, param_distributions = param_grid, n_iter = 30, cv = 3, verbose=2, random_state=42, n_jobs = -1)
clf_random.fit(X_resampled, y_resampled)
```

```python
clf_random.best_params_
```

Cross-validation is a technique for evaluating a machine learning model and testing its performance. CV is commonly used in applied ML tasks. It helps to compare and select an appropriate model for the specific predictive modeling problem.

CV is easy to understand, easy to implement, and it tends to have a lower bias than other methods used to count the modelâ€™s efficiency scores. All this makes cross-validation a powerful tool for selecting the best model for the specific task.

Types of CV: <br> 
1. Hold-out cross-validation 
2. k-Fold cross-validation: all data is used for training  
3. Leave-one-out cross-validation: k is equal to n where n is the number of samples in the dataset 
4. Stratified k-Fold cross-validation: variation of the standard k-Fold CV technique which is designed to be effective in such cases of target imbalance
5. Time-series cross-validation: traditional cross-validation techniques donâ€™t work on sequential data such as time-series by nature. 

Here we already balanced the classes with oversampling so we can safely use k-Fold CV (instead of stratified CV).

```python
from sklearn.model_selection import KFold


# Define the LightGBM classifier with the best parameters found from the randomized search
clf = lgb.LGBMClassifier(**clf_random.best_params_)

# Define the KFold cross-validation object with 5 folds
kf = KFold(n_splits=10, shuffle=True, random_state=42)

# Create an empty array to store the cross-validation scores
scores = np.zeros(10)

# Loop over the folds of the KFold cross-validation object
for i, (train_index, val_index) in enumerate(kf.split(X_resampled, y_resampled)):
    
    # Split the resampled data into training and testing sets for this fold
    X_train_, X_test_ = X_resampled.iloc[train_index], X_resampled.iloc[val_index]
    y_train_, y_test_ = y_resampled.iloc[train_index], y_resampled.iloc[val_index]
    
    # Fit the LightGBM classifier on the training data
    clf.fit(X_train_, y_train_)
    
    # Evaluate the performance of the classifier on the testing data
    scores[i] = clf.score(X_test_, y_test_)
    print(f'Step {i+1}/10 done...')

# Print the mean and standard deviation of the cross-validation scores
print("Cross-validation scores: {}".format(scores))
print("Mean score: {:.3f} (+/- {:.3f})".format(scores.mean(), scores.std() * 2))
```

The accuracy seems pretty stable at a high level and the standard deviation is low enough. 

```python
# Get the best hyperparameters
best_clf = clf_random.best_estimator_

# Predict the target using the best classifier
y_pred = best_clf.predict(X_test)
```

### Results interpretation

```python
import seaborn as sn
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.metrics import classification_report


# Calculate the evaluation metrics 
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred, average='macro')
recall = recall_score(y_test, y_pred, average='macro')
f1 = f1_score(y_test, y_pred, average='macro')

# Print the evaluation metrics
print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("F1 Score:", f1, '\n')

# Print classification report
print("Classification Report:\n\n", classification_report(y_test , y_pred, digits=3))
```

All our metrics have improved by adding Opensea data to the equation.
